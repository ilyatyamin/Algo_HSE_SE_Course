## 1. Разработайте программу для вычисления приблизительного значения числа т по алгоритму, представленному выше. Оформите случайную генерацию координат точек в виде отдельной функции.

```python
class Point(object):
  def __init__(self, x, y):
    self.x = x
    self.y = y
```

```python
def generate_point():
  x = random.uniform(-1, 1)
  y = random.uniform(-1, 1)

  return Point(x, y)

def generate_mas_points(count : int):
  points = []
  for i in range(count):
    points.append(generate_point())
  return points
```

```python
def pi_approximate(count : int):
  points = generate_mas_points(count)

  # Посчитаем сколько из них попадает в круг
  into_circle = 0
  for point in points:
    if (point.x**2 + point.y**2 <= 1):
      into_circle += 1

  return (4*into_circle) / count
```

```python
print(pi_approximate(100))
```

## 2. Проведите ряд экспериментов для оценки точности вычисления значения числа в зависимости от разного количества случайно сгенерированных точек N, которое меняется от 100 до 5000 с шагом 100. Представьте результаты проведенных экспериментов в следующем виде:


*   a) График 1, который отображает, как изменяется приближенное значение числа п, полученное в разработанной программе, в зависимости от общего числа точек N.
*   b. График 2, который отображает, как изменяется относительное отклонение (в %) приближенного значения числа п от точного в зависимости от общего числа точек N.
* c. Исходные данные, использованные для построения графиков. Для удобства представления, графики можно разделить на несколько.


### блок a

```python
nums = [ num for num in range(100, 5001, 100) ]
results_a = [ pi_approximate(num) for num in nums ]
```

```python
fig, ax = plt.subplots()
ax.scatter(nums, results_a)
plt.axline( (0, math.pi), (5000, math.pi), color='r')
plt.xlabel('Количество точек')
plt.ylabel('Значение числа pi')
plt.title('Задание 2a')
```

Здесь горизонтальной красной прямой показано реальное значение числа $\pi$

### блок B

```python
results_B = [ abs((math.pi - value) / math.pi)*100 for value in results_a ]
```

```python
fig, ax = plt.subplots()
ax.scatter(nums, results_B)
plt.xlabel('Количество точек')
plt.ylabel('Относительное отклонение в %')
plt.title('Задание 2b')
```

### блок C

```python
dataset = pd.DataFrame({"N" : nums, "pi" : results_a, "отклонение" : results_B})
display(dataset)
```
