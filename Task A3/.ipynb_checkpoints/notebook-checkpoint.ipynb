```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
quick1 = pd.read_csv('quick1.txt', sep=';', names=[ 'кол-во данных', 'время'])
quick2 = pd.read_csv('quick2.txt', sep=';', names=['кол-во данных', 'время'])
quick3 = pd.read_csv('quick3.txt', sep=';', names=[ 'кол-во данных', 'время'])

heapQuick1 = pd.read_csv('quickHeap1.txt', sep=';', names=['кол-во данных', 'граница перехода в Heap', 'время'])
heapQuick2 = pd.read_csv('quickHeap2.txt', sep=';', names=['кол-во данных', 'граница перехода в Heap', 'время'])
heapQuick3 = pd.read_csv('quickHeap3.txt', sep=';', names=['кол-во данных', 'граница перехода в Heap', 'время'])
```

```python
# Берем средние значения по всему
quick1 = quick1.groupby('кол-во данных', as_index = False).mean()
quick2 = quick2.groupby('кол-во данных', as_index = False).mean()
quick3 = quick3.groupby('кол-во данных', as_index = False).mean()

heapQuick1 = heapQuick1.groupby(['кол-во данных', 'граница перехода в Heap'], as_index = False).mean()
heapQuick2 = heapQuick2.groupby(['кол-во данных', 'граница перехода в Heap'], as_index = False).mean()
heapQuick3 = heapQuick3.groupby(['кол-во данных', 'граница перехода в Heap'], as_index = False).mean()
```

```python
fig, ax = plt.subplots()
ax.plot(quick1['кол-во данных'], quick1['время'])
plt.xlabel('Количество данных')
plt.ylabel('Время в миллисекундах')
plt.title('Тестовый набор №1: полностью случайный (quickSort)')
```

```python
fig, ax = plt.subplots()
ax.plot(quick2['кол-во данных'], quick2['время'])
plt.xlabel('Количество данных')
plt.ylabel('Время в миллисекундах')
plt.title('Тестовый набор №2: в обратном порядке (quickSort)')
```

```python
fig, ax = plt.subplots()
ax.plot(quick3['кол-во данных'], quick3['время'])
plt.xlabel('Количество данных')
plt.ylabel('Время в миллисекундах')
plt.title('Тестовый набор №3: практически sorted (quickSort)')
```
