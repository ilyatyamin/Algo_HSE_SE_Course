```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
merge1 = pd.read_csv('test_merge_1.txt', sep=';', names=[ 'кол-во данных', 'время'])
merge2 = pd.read_csv('test_merge_2.txt', sep=';', names=['кол-во данных', 'время'])
merge3 = pd.read_csv('test_merge_3.txt', sep=';', names=[ 'кол-во данных', 'время'])

insMerge1 = pd.read_csv('test_mergeInsertion_1.txt', sep=';', names=['граница перехода в Insertion', 'время'])
insMerge2 = pd.read_csv('test_mergeInsertion_2.txt', sep=';', names=[ 'граница перехода в Insertion', 'время'])
insMerge3 = pd.read_csv('test_mergeInsertion_3.txt', sep=';', names=['граница перехода в Insertion', 'время'])
```

```python
fig, ax = plt.subplots()
ax.plot(merge1['кол-во данных'], merge1['время'])
plt.xlabel('Количество данных')
plt.ylabel('Время в миллисекундах')
plt.title('Тестовый набор №1: полностью случайный')
```

```python
fig, ax = plt.subplots()
ax.plot(merge2['кол-во данных'], merge2['время'])
plt.xlabel('Количество данных')
plt.ylabel('Время в миллисекундах')
plt.title('Тестовый набор №2: отсортированный в обратном порядке')
```

```python
fig, ax = plt.subplots()
ax.plot(merge3['кол-во данных'], merge3['время'])
plt.xlabel('Количество данных')
plt.ylabel('Время в миллисекундах')
plt.title('Тестовый набор №3: практически отсортированный')
```

```python
fig, ax = plt.subplots()
ax.plot(insMerge1['граница перехода в Insertion'], insMerge1['время'])
plt.xlabel('Граница перехода в Insertion')
plt.ylabel('Время в миллисекундах')
plt.title('Тестовый набор №1: полностью случайный')
```

```python
fig, ax = plt.subplots()
ax.plot(insMerge2['граница перехода в Insertion'], insMerge2['время'])
plt.xlabel('Граница перехода в Insertion')
plt.ylabel('Время в миллисекундах')
plt.title('Тестовый набор №2: в обратном порядке')
```

```python
fig, ax = plt.subplots()
ax.plot(insMerge3['граница перехода в Insertion'], insMerge3['время'])
plt.xlabel('Граница перехода в Insertion')
plt.ylabel('Время в миллисекундах')
plt.title('Тестовый набор №3: практически отсортированный')
```

```python
# Сравним MergeSort на одном графике
fig, ax = plt.subplots()
ax.plot(merge1['кол-во данных'], merge1['время'], color = 'r')
ax.plot(merge2['кол-во данных'], merge2['время'], color = 'b')
ax.plot(merge3['кол-во данных'], merge3['время'], color = 'g')
plt.xlabel('Количество данных')
plt.ylabel('Время в миллисекундах')
plt.title('Все тестовые наборы')
```

```python
fig, ax = plt.subplots()
ax.plot(insMerge1['граница перехода в Insertion'], insMerge1['время'], color = 'r')
ax.plot(insMerge2['граница перехода в Insertion'], insMerge2['время'], color = 'g')
ax.plot(insMerge3['граница перехода в Insertion'], insMerge3['время'], color = 'b')
plt.xlabel('Граница перехода в Insertion')
plt.ylabel('Время в миллисекундах')
plt.title('Гибридный алгоритм (красный - 1 набор, зел - 2, синий - 3)')
```
