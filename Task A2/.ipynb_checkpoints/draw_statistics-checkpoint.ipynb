```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
merge1 = pd.read_csv('test_merge_1.txt', sep=';', names=[ 'кол-во данных', 'время'])
merge2 = pd.read_csv('test_merge_2.txt', sep=';', names=['кол-во данных', 'время'])
merge3 = pd.read_csv('test_merge_3.txt', sep=';', names=[ 'кол-во данных', 'время'])

insMerge1 = pd.read_csv('test_mergeInsertion_1.txt', sep=';', names=['кол-во данных', 'граница перехода в Insertion', 'время'])
insMerge2 = pd.read_csv('test_mergeInsertion_2.txt', sep=';', names=['кол-во данных', 'граница перехода в Insertion', 'время'])
insMerge3 = pd.read_csv('test_mergeInsertion_3.txt', sep=';', names=['кол-во данных', 'граница перехода в Insertion', 'время'])
```

```python
# Берем средние значения по всему
merge1 = merge1.groupby('кол-во данных', as_index = False).mean()
merge2 = merge2.groupby('кол-во данных', as_index = False).mean()
merge3 = merge3.groupby('кол-во данных', as_index = False).mean()

insMerge1 = insMerge1.groupby(['кол-во данных', 'граница перехода в Insertion'], as_index = False).mean()
insMerge2 = insMerge2.groupby(['кол-во данных', 'граница перехода в Insertion'], as_index = False).mean()
insMerge3 = insMerge3.groupby(['кол-во данных', 'граница перехода в Insertion'], as_index = False).mean()
```

```python
fig, ax = plt.subplots()
ax.plot(merge1['кол-во данных'], merge1['время'])
plt.xlabel('Количество данных')
plt.ylabel('Время в миллисекундах')
plt.title('Тестовый набор №1: полностью случайный')
```

```python
fig, ax = plt.subplots()
ax.plot(merge2['кол-во данных'], merge2['время'])
plt.xlabel('Количество данных')
plt.ylabel('Время в миллисекундах')
plt.title('Тестовый набор №2: отсортированный в обратном порядке')
```
